```python
"""
Plotting V=I*R
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

voltage=np.linspace(0,100,50)
resistance=30
current=voltage/resistance

plt.plot(voltage,current,label="V=I*R")
plt.legend()
plt.title("Ohm's Law")
```

```python
'''Customizing the colours '''
plt.subplot(1,2,1)
plt.plot(voltage,current,'r')
plt.show()

plt.subplot(1,2,2)
plt.plot(voltage,current,'b')
plt.show()

plt.subplot(1,1,1)
plt.plot(voltage,current,'g')
plt.show()
```

```python
''' Working With Subplots'''
x=np.linspace(1,10,100)
y=np.log(x)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("y=log(x)")


plt.subplot(1,2,2)
plt.plot(x,np.exp(y))
plt.title("y=x")
```

Types Of Commonly Used Plots

```python
df=pd.read_csv('insurance.csv')
df.head()
```

Boxplot

```python
plt.figure(figsize=(5,5))
plt.boxplot(df.expenses)
```

Histogram

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(df.expenses,bins=100)
plt.title('Expenses')

plt.subplot(1,2,2)
plt.hist(df.bmi,bins=50)
plt.title('BMI')

```

Scatterplot

```python
plt.scatter(df.bmi,df.expenses,c='k')
```

Bar Charts

```python
df_avg_age_prem=df.groupby('age')['expenses'].mean()
plt.bar(df_avg_age_prem.index,df_avg_age_prem)
```
