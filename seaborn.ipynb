```python
import seaborn as sns
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('insurance.csv')
df.head()
```

```python
df_avg_exp_age=df.groupby('age')['expenses'].mean().to_frame().reset_index()
sns.lineplot(x='age',y='expenses',data=df_avg_exp_age)
```

```python
df_prem_age=df.groupby(['region','sex'])['expenses'].mean().to_frame().reset_index()
sns.barplot(x='region',y='expenses',hue='sex',data=df_prem_age)
```

```python
sns.displot(x='expenses',data=df,kind='kde')
```

```python
sns.boxplot(x='region',y='bmi',data=df)
```

```python
sns.violinplot(x='bmi',y='region',data=df)
```

```python
sns.relplot(x='age',y='bmi',kind='scatter',data=df,hue='sex')
```

```python

sns.relplot(x='age',y='expenses',size='bmi',data=df,alpha=0.5)
```

```python
sns.relplot(x='age',y='expenses',size='bmi',row='sex',col='region',data=df,alpha=0.5)
```

```python
sns.catplot(x='region',y='bmi',data=df)
```

```python
sns.catplot(x='region',y='bmi',data=df,kind='point')
```

```python
sns.pairplot(df)
```
